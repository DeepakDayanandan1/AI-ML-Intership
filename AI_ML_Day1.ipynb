```python
movies = [{"title":"Inception","year":2010,"ratings":[8,9,10,9]},
          {"title":"Titanic","year":1997,"ratings":[9,8,9,10]},
           {"title":"The Matrix","year":1999,"ratings":[10,9,10,10]},
            {"title":"Interstellar","year":2014,"ratings":[9,8,10,9]}]
print(movies)
```

```python
movies.append({"title":"Deadpool","year":2024,"ratings":[9,10,10,9]})
print(movies)
```

```python
for movie in movies:
    if movie["title"] == "Inception":
        movie["ratings"].extend([9, 10, 8])
print(movies[0])
```

```python
for movie in movies:
  rating= sum(movie["ratings"])/len(movie["ratings"])
  print("Movie title:",movie["title"],"Average rating:",rating)
```

```python
for movie in movies:
    if movie["title"] == "Inception":
        movie["title"] = "Cobra Kai"
print(movies)
```

```python
highest_rated = max(movies, key=lambda movie: movie["ratings"])
print("Highest rated movie:", highest_rated['title'], "with rating", highest_rated['ratings'])
```

```python
movies.pop(1)
print(movies)
```

```python
for movie in movies:
  rating= sum(movie["ratings"])/len(movie["ratings"])
  if rating >= 9:
    print("Movie:",movie["title"])
```

```python
marks = [90, 40, 75, 52, 56]

#def calculate_average(marks_list):
 #   return sum(marks_list)/len(marks_list)

def assign_grade(mark):
    if mark >= 90:
        return 'A'
    elif mark >= 80:
        return 'B'
    elif mark >= 70:
        return 'C'
    elif mark >= 60:
        return 'D'
    else:
        return 'F'

#average = calculate_average(marks)

grades = [assign_grade(mark) for mark in marks]

print("Student's Marks:", marks)
print("Grades:", grades)
#print("Average Mark:", average)
```
