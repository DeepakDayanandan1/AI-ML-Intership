**MATPLOTLIB**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,3,4,5,10,11,23,15,9]

avg1 = np.mean(x)
avg2 = max(y)

plt.figure(figsize=(5,4))
plt.plot(x,y,color='green',linestyle='dashed',marker='*')

plt.axhline(avg2,color='red')
plt.axvline(avg1,color='red')

plt.grid()
plt.title("Line Chart")
plt.xlabel("x values")
plt.xlabel("y values")
plt.show()
```

```python
plt.scatter(x,y,marker="*")
plt.show()
```

```python
products = ['Books','Bottles','Shoes','Bags','Clothes','Cosmetics']
sales = [2500,1000,8000,12000,9000,4000]
plt.bar(products,sales)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.barh(products,sales)
```

```python
plt.pie(sales,labels=products,shadow=True,autopct='%1.1f%%')
plt.show()
```

```python
days = [1,2,3,4,5]
reading = [1,2,2.5,2,3]
studying = [3,4,4,5,3]
playing = [7,8,9,6,8]

plt.stackplot(days,studying,reading,playing)
plt.legend(['study hours','for reading'])
plt.show()
```

```python
sales_data = [np.random.randint(100,1000) for i in range(30)]
sales_data
```

```python
plt.hist(sales_data,bins=10,edgecolor='white')
plt.show()
```

```python
fig,axs = plt.subplots(2,2,figsize=(10,8))
axs[0,0].bar(products,sales)
axs[0,0].set_title("Products vs Sales")

axs[0,1].pie(sales,labels=products,shadow=True,autopct='%1.1f%%')
axs[0,1].set_title("Sales Contribution")

axs[1,0].stackplot(days,studying,reading,playing)
axs[1,0].set_title("Hours Spend")

axs[1,1].hist(sales_data,bins=10)
axs[1,1].set_title("Daily Stats")

plt.suptitle("MULTIPLE CHARTS DIAGRAM")
plt.show()
```

```python
#catplot
import seaborn as sns

exercise = sns.load_dataset("exercise")
sns.catplot(x="time", y="pulse",hue="kind",data=exercise)
```

```python
#swarmplot
data = sns.load_dataset("tips")
sns.swarmplot(x ="day", y = "total_bill", data = data, size = 5)
plt.show()
```

```python
#violinplot
sns.set(style = 'whitegrid')
fmri = sns.load_dataset("fmri")

sns.violinplot(x ="timepoint", y ="signal", data = fmri)
```

```python
#stripplot
sns.set(style = 'whitegrid')
tip = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=tip)
plt.show()
```
